```python
#6. Univariate Analysis
# We examine the distribution of individual variables such as Age and Fare.
```

```python
#Univariate Analysis (Age & Fare Distributions)

#Age Distribution
plt.hist(df["Age"].dropna(), bins=20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Fare Distribution
plt.hist(df["Fare"], bins=20)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
#Observation:

#- Most passengers were between 20 and 40 years old.
#- Fare distribution is right-skewed, meaning most passengers paid lower fares, while few paid very high fares.
```

```python
# Bivariate Analysis

#We analyze the relationship between survival and other variables such as Gender and Passenger Class.
```

```python
#Bivariate Analysis (Survival by Gender)

survival_gender = df.groupby("Sex")["Survived"].mean()

print("\nSurvival Rate by Gender:")
print(survival_gender)

survival_gender.plot(kind="bar")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

```

```python
#Observation:

#Female passengers had a significantly higher survival rate than male passengers.  
#Gender appears to be a strong factor influencing survival.
```

```python
#Bivariate Analysis (Survival by Passenger Class)
survival_class = df.groupby("Pclass")["Survived"].mean()

print("\nSurvival Rate by Passenger Class:")
print(survival_class)

survival_class.plot(kind="bar")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#Observation:

#Passengers in First Class had the highest survival rate, while Third Class passengers had the lowest.  
#This suggests that passenger class influenced survival probability.
```

```python
#Correlation Matrix

#We examine correlations between numeric variables to identify relationships.
```

```python
#Correlation Matrix
correlation = df.corr(numeric_only=True)

print("\nCorrelation Matrix:")
print(correlation)

plt.matshow(correlation)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()

print("\nCorrelation with Survival:")
print(correlation["Survived"].sort_values(ascending=False))
```

```python
#Observation:

#- Passenger class has a negative correlation with survival.
#- Fare shows a positive correlation with survival.
#- Some variables have weak relationships.
```

```python
# Data Cleaning

#We handle missing values and remove columns with too many missing entries.
```

```python
# Fill missing Age
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop("Cabin", axis=1, errors="ignore")
print("\nMissing Values After Cleaning:")
print(df.isna().sum())
```
